```python
PROMPT_REACT = """
VocĂȘ funciona em um ciclo de Pensamento, AĂ§ĂŁo, Pausa e ObservaĂ§ĂŁo.
Ao final do ciclo, vocĂȘ fornece uma Resposta.
Use "Pensamento" para descrever seu raciocĂ­nio.
Use "AĂ§ĂŁo" para executar ferramentas - e entĂŁo retorne "PAUSA".
A "ObservaĂ§ĂŁo" serĂĄ o resultado da aĂ§ĂŁo executada.
AĂ§Ă”es disponĂ­veis:
  - consultar_estoque: retorna a quantidade disponĂ­vel de um item no inventĂĄrio (ex: "consultar_estoque: teclado")
  - consultar_preco_produto: retorna o preĂ§o unitĂĄrio de um produto (ex: "consultar_preco_produto: mouse gamer")

Exemplo:
Pergunta: Quantos monitores temos em estoque?
Pensamento: Devo consultar a aĂ§ĂŁo consultar_estoque para saber a quantidade de monitores.
AĂ§ĂŁo: consultar_estoque: monitor
PAUSA

ObservaĂ§ĂŁo: Temos 75 monitores em estoque.
Resposta: HĂĄ 75 monitores em estoque.
""".strip()
```

```python
def consultar_estoque(item: str) -> str:
    """
    Simula a consulta de estoque de um item no inventĂĄrio.
    """
    item = item.lower()
    estoque = {
        "monitor": 75,
        "teclado": 120,
        "mouse gamer": 80,
        "webcam": 40,
        "headset": 60,
        "impressora": 15
    }

    if item in estoque:
        return f"Temos {estoque[item]} {item}s em estoque."
    else:
        return f"Item '{item}' nĂŁo encontrado no inventĂĄrio."

def consultar_preco_produto(produto: str) -> str:
    """
    Simula a consulta do preĂ§o unitĂĄrio de um produto.
    """
    produto = produto.lower()
    precos = {
        "monitor": 999.90,
        "teclado": 150.00,
        "mouse gamer": 99.50,
        "webcam": 120.00,
        "headset": 180.00,
        "impressora": 750.00
    }

    if produto in precos:
        return f"O preĂ§o de um(a) {produto} Ă© R$ {precos[produto]:.2f}."
    else:
        return f"Produto '{produto}' nĂŁo encontrado na lista de preĂ§os."

```

```python
def run_react_agent(pergunta: str, max_iterations: int = 5) -> str:
    model = genai.GenerativeModel('gemini-2.5-flash')
    chat = model.start_chat(history=[])

    chat.send_message(PROMPT_REACT)

    current_prompt = pergunta

    for i in range(max_iterations):
        response = chat.send_message(current_prompt)
        response_text = response.text.strip()

        print(f"--- IteraĂ§ĂŁo {i+1} ---")
        print(f"Modelo pensou/respondeu:\n{response_text}\n")

        if response_text.startswith("Resposta:"):
            return response_text.replace("Resposta:", "").strip()

        match = re.search(r"AĂ§ĂŁo:\s*(\w+):\s*(.*)", response_text)

        if match:
            action_name = match.group(1).strip()
            action_arg = match.group(2).strip()

            observacao = ""
            if action_name == "consultar_estoque":
                observacao = consultar_estoque(action_arg)
            elif action_name == "consultar_preco_produto":
                observacao = consultar_preco_produto(action_arg)
            else:
                observacao = f"Erro: AĂ§ĂŁo '{action_name}' desconhecida."

            current_prompt = f"ObservaĂ§ĂŁo: {observacao}\nResposta:"

            print(f"Executou aĂ§ĂŁo: {action_name}({action_arg})")
            print(f"ObservaĂ§ĂŁo: {observacao}\n")

        else:
            return f"Erro: O agente nĂŁo conseguiu extrair uma AĂ§ĂŁo ou Resposta final apĂłs {i+1} iteraĂ§Ă”es. Ăltima resposta: {response_text}"

    return "Erro: Limite mĂĄximo de iteraĂ§Ă”es atingido sem uma resposta final."
```

```python
def consultar_estoque(item: str) -> str: 
    """
    Simula a consulta de estoque de um item no inventĂĄrio.
    """
    item = item.lower().strip()
    estoque = {
        "monitor": 75,
        "teclado": 120,
        "mouse gamer": 80,
        "webcam": 40,
        "headset": 60,
        "impressora": 15
    }

    if item in estoque:
        return f"Temos {estoque[item]} {item}s em estoque."
    else:
        return f"Item '{item}' nĂŁo encontrado no inventĂĄrio."

def consultar_preco_produto(produto: str) -> str: 
    """
    Simula a consulta do preĂ§o unitĂĄrio de um produto.
    """
    produto = produto.lower().strip()
    precos = {
        "monitor": 999.90,
        "teclado": 150.00,
        "mouse gamer": 99.50,
        "webcam": 120.00,
        "headset": 180.00,
        "impressora": 750.00
    }

    if produto in precos:
        return f"O preĂ§o de um(a) {produto} Ă© R$ {precos[produto]:.2f}."
    else:
        return f"Produto '{produto}' nĂŁo encontrado na lista de preĂ§os."

def ferramenta_encontrar_produto_mais_caro() -> str: 
    """
    Retorna o nome e o preĂ§o do produto mais caro no inventĂĄrio.
    Esta funĂ§ĂŁo nĂŁo precisa de argumentos.
    """
    
    precos_do_inventario = {
        "monitor": 999.90,
        "teclado": 150.00,
        "mouse gamer": 99.50,
        "webcam": 120.00,
        "headset": 180.00,
        "impressora": 750.00
    }

    if not precos_do_inventario: 
        return "Nenhum produto encontrado na lista de preĂ§os para comparaĂ§ĂŁo."

    
    nome_produto_mais_caro = max(precos_do_inventario, key=precos_do_inventario.get)
    valor_produto_mais_caro = precos_do_inventario[nome_produto_mais_caro]

    return f"O produto mais caro Ă© o(a) {nome_produto_mais_caro} com preĂ§o de R$ {valor_produto_mais_caro:.2f}." 


```

```python
PROMPT_REACT = """
VocĂȘ funciona em um ciclo de Pensamento, AĂ§ĂŁo, Pausa e ObservaĂ§ĂŁo.
Ao final do ciclo, vocĂȘ fornece uma Resposta.
Use "Pensamento" para descrever seu raciocĂ­nio.
Use "AĂ§ĂŁo" para executar ferramentas - e entĂŁo retorne "PAUSA".
A "ObservaĂ§ĂŁo" serĂĄ o resultado da aĂ§ĂŁo executada.
AĂ§Ă”es disponĂ­veis:
  - consultar_estoque: retorna a quantidade disponĂ­vel de um item no inventĂĄrio (ex: "consultar_estoque: teclado")
  - consultar_preco_produto: retorna o preĂ§o unitĂĄrio de um produto (ex: "consultar_preco_produto: mouse gamer")
  - encontrar_produto_mais_caro: retorna o nome e o preĂ§o do produto mais caro no inventĂĄrio (nĂŁo requer argumentos)

Exemplo:
Pergunta: Quantos monitores temos em estoque?
Pensamento: Devo consultar a aĂ§ĂŁo consultar_estoque para saber a quantidade de monitores.
AĂ§ĂŁo: consultar_estoque: monitor
PAUSA

ObservaĂ§ĂŁo: Temos 75 monitores em estoque.
Resposta: HĂĄ 75 monitores em estoque.

Exemplo:
Pergunta: Qual Ă© o produto mais caro?
Pensamento: Preciso usar a aĂ§ĂŁo encontrar_produto_mais_caro para descobrir qual produto tem o maior preĂ§o.
AĂ§ĂŁo: encontrar_produto_mais_caro
PAUSA

ObservaĂ§ĂŁo: O produto mais caro Ă© o(a) monitor com preĂ§o de R$ 999.90.
Resposta: O produto mais caro Ă© o(a) monitor com preĂ§o de R$ 999.90.
""".strip()
```

```python
def run_react_agent(pergunta: str, max_iterations: int = 5) -> str:
    """
    Executa o ciclo ReAct para uma dada pergunta usando o modelo Gemini.
    """

    model = genai.GenerativeModel('gemini-2.5-flash')

    chat = model.start_chat(history=[])
    chat.send_message(PROMPT_REACT)

    current_prompt = pergunta

    for i in range(max_iterations):
        response = chat.send_message(current_prompt)
        response_text = response.text.strip()

        print(f"\n--- IteraĂ§ĂŁo {i+1} ---")
        print(f"Modelo pensou/respondeu:\n{response_text}\n")

        
        response_match_final = re.search(r"Resposta:\s*(.*)", response_text, re.DOTALL)
        if response_match_final:
            return response_match_final.group(1).strip()

        
        match = re.search(r"AĂ§ĂŁo:\s*(\w+)(?::\s*([^\n]*))?", response_text)

        if match:
            action_name = match.group(1).strip()

            action_arg = match.group(2).strip() if match.group(2) is not None else ""

            observacao_da_acao = ""

            if action_name == "consultar_estoque":
                observacao_da_acao = consultar_estoque(action_arg)
            elif action_name == "consultar_preco_produto":
                observacao_da_acao = consultar_preco_produto(action_arg)
            elif action_name == "encontrar_produto_mais_caro": 
                observacao_da_acao = ferramenta_encontrar_produto_mais_caro()
            else:
                observacao_da_acao = f"Erro: AĂ§ĂŁo '{action_name}' desconhecida. Verifique o prompt ou a implementaĂ§ĂŁo da ferramenta."

            # Simplificamos o prompt de ObservaĂ§ĂŁo
            current_prompt = f"ObservaĂ§ĂŁo: {observacao_da_acao}"

            print(f"Executou aĂ§ĂŁo: {action_name} com argumento '{action_arg}'")
            print(f"ObservaĂ§ĂŁo: {observacao_da_acao}\n")

        else:
            
            return f"Erro: O agente nĂŁo conseguiu extrair uma AĂ§ĂŁo ou Resposta final apĂłs {i+1} iteraĂ§Ă”es. Ăltima resposta do modelo: {response_text}"

    
    return "Erro: Limite mĂĄximo de iteraĂ§Ă”es atingido sem uma resposta final do agente."
```

```python
def run_react_agent_with_history(pergunta: str, max_iterations: int = 5) -> tuple[str, list]: 
    model = genai.GenerativeModel('gemini-2.5-flash') 

    chat = model.start_chat(history=[])
    chat.send_message(PROMPT_REACT) 

    current_prompt = pergunta

    for i in range(max_iterations):
        response = chat.send_message(current_prompt)
        response_text = response.text.strip()

        print(f"\n--- IteraĂ§ĂŁo {i+1} ---")
        print(f"Modelo pensou/respondeu:\n{response_text}\n")

        response_match_final = re.search(r"Resposta:\s*(.*)", response_text, re.DOTALL)
        if response_match_final:
            return response_match_final.group(1).strip(), chat.history

        match = re.search(r"AĂ§ĂŁo:\s*(\w+)(?::\s*([^\n]*))?", response_text)

        if match:
            action_name = match.group(1).strip()
            action_arg = match.group(2).strip() if match.group(2) is not None else ""

            observacao_da_acao = ""

            if action_name == "consultar_estoque":
                observacao_da_acao = consultar_estoque(action_arg)
            elif action_name == "consultar_preco_produto":
                observacao_da_acao = consultar_preco_produto(action_arg)
            elif action_name == "encontrar_produto_mais_caro":
                observacao_da_acao = ferramenta_encontrar_produto_mais_caro()
            else:
                observacao_da_acao = f"Erro: AĂ§ĂŁo '{action_name}' desconhecida. Verifique o prompt ou a implementaĂ§ĂŁo da ferramenta."

            current_prompt = f"ObservaĂ§ĂŁo: {observacao_da_acao}"

            print(f"Executou aĂ§ĂŁo: {action_name} com argumento '{action_arg}'")
            print(f"ObservaĂ§ĂŁo: {observacao_da_acao}\n")

        else:
            return f"Erro: O agente nĂŁo conseguiu extrair uma AĂ§ĂŁo ou Resposta final apĂłs {i+1} iteraĂ§Ă”es. Ăltima resposta do modelo: {response_text}", chat.history

    return "Erro: Limite mĂĄximo de iteraĂ§Ă”es atingido sem uma resposta final do agente.", chat.history
```

```python
def ferramenta_calcular_valor_total_lista(lista_itens: str) -> str:
    """
    Calcula o valor total de uma lista de itens de compra.
    Recebe uma string com itens separados por vĂ­rgula (ex: "teclado, mouse gamer, monitor").
    """
    precos_do_inventario = { 
        "monitor": 999.90,
        "teclado": 150.00,
        "mouse gamer": 99.50,
        "webcam": 120.00,
        "headset": 180.00,
        "impressora": 750.00
    }

    itens_processados = [item.strip().lower() for item in lista_itens.split(',')]

    valor_total = 0.0
    itens_nao_encontrados = []


    for item in itens_processados: 
        if item in precos_do_inventario:
            valor_total += precos_do_inventario[item]
        else:
            itens_nao_encontrados.append(item)

    resposta = f"O valor total dos itens encontrados Ă© R$ {valor_total:.2f}."
    if itens_nao_encontrados:
        resposta += f" Os seguintes itens nĂŁo foram encontrados e nĂŁo foram incluĂ­dos no cĂĄlculo: {', '.join(itens_nao_encontrados)}."

    return resposta
```
